```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import pickle

# Load the car data.
df = pd.read_csv("new_cars.csv")

le = LabelEncoder()

Name_encode = le.fit_transform(df['Name_of_Cars'])
df.drop('Name_of_Cars',axis=1,inplace=True)
df['Name_of_Cars'] = Name_encode

Year_encode = le.fit_transform(df['Year'])
df.drop('Year',axis=1,inplace=True)
df['Year'] = Year_encode

Miles_encode = le.fit_transform(df['Miles'])
df.drop('Miles',axis=1,inplace=True)
df['Miles'] = Miles_encode

# Create the features and target variables.
features = df.drop("Price", axis=1)
target = df["Price"]

# Split the data into a training set and a test set.
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25)

# Create the model.
model = LinearRegression()

# Train the model.
model.fit(X_train, y_train)

# Evaluate the model.
score = model.score(X_test, y_test)
print("Score:", score)

# Make predictions.
predictions = model.predict(X_test)

# Print the predictions.
print("Predictions:", predictions)

filename = 'car_price_prediction_model1.pkl'  # Choose a filename for the model
with open(filename, 'wb') as file:
    pickle.dump(model, file)
```
